```python
import requests

# URL for login and data access
login_url = 'https://cogitate-data.ae.mpg.de/login'
data_url = 'https://cogitate-data.ae.mpg.de/download/data_bundle.zip'

# Your login credentials
payload = {
    'username': 'your_username',
    'password': 'your_password'
}

# Start a session
session = requests.Session()

# Perform login
response = session.post(login_url, data=payload)

# Check if login was successful
if response.status_code == 200:
    print('Login successful!')

    # Download the data
    data_response = session.get(data_url, stream=True)

    # Save the data to a file
    with open('/content/data_bundle.zip', 'wb') as file:
        for chunk in data_response.iter_content(chunk_size=8192):
            if chunk:
                print(chunk)
                file.write(chunk)
    print('Data downloaded and saved successfully!')
else:
    print('Login failed. Please check your credentials.')
```
